```python
import pandas as pd
from sqlalchemy import create_engine, text

DB_CONFIG = {
    "host": "100.105.75.47",
    "port": 5432,
    "dbname": "postgres",
    "user": "postgres",
    "password": "
",
}

engine = create_engine(
    f"postgresql+psycopg2://{DB_CONFIG['user']}:{DB_CONFIG['password']}"
    f"@{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['dbname']}",
    pool_size=1,
    max_overflow=0,
    pool_pre_ping=True,
)

# -------------------------
# 유틸: end_day(YYYYMMDD) -> prev month(YYYYMM)
# 예) 20260129 -> 202512
# -------------------------
def prev_month_yyyymm(end_day_str: str) -> str:
    s = str(end_day_str)
    y = int(s[:4])
    m = int(s[4:6])
    if m == 1:
        return f"{y-1}12"
    return f"{y}{m-1:02d}"

# -------------------------
# 유틸: remark 정규화 (Non-PD 매칭 이슈 방지)
#  - 공백/언더스코어 제거, 대문자화 후 PD / Non-PD로 통일
# -------------------------
def normalize_remark(x: str) -> str | None:
    if x is None:
        return None
    s = str(x).strip()
    u = s.upper().replace(" ", "").replace("_", "")
    # 대시가 다른 문자(–, —)일 수 있으니 통일
    u = u.replace("–", "-").replace("—", "-")
    if u == "PD":
        return "PD"
    if u in ("NON-PD", "NONPD"):
        return "Non-PD"
    return None

# -------------------------
# (A) BadFile 데이터 조회
# -------------------------
SRC_SCHEMA = "a1_fct_vision_testlog_txt_processing_history"
SRC_TABLE  = "fct_vision_testlog_txt_processing_history"

stations = ("FCT1", "FCT2", "FCT3", "FCT4")
goodorbad_value = "BadFile"

sql_badfile = f"""
SELECT
    barcode_information,
    station,
    end_day,
    end_time,
    remark,
    result,
    goodorbad,
    file_path
FROM {SRC_SCHEMA}.{SRC_TABLE}
WHERE station = ANY(:stations)
  AND goodorbad = :goodorbad
ORDER BY end_day DESC, end_time DESC
"""

with engine.connect() as conn:
    df = pd.read_sql_query(
        text(sql_badfile),
        conn,
        params={"stations": list(stations), "goodorbad": goodorbad_value},
    )

# prev month 컬럼 생성
df["month"] = df["end_day"].astype(str).apply(prev_month_yyyymm)

# remark 정규화 컬럼(조인키용)
df["remark_norm"] = df["remark"].apply(normalize_remark)

# -------------------------
# (B) op_ct 테이블 위치 자동 탐색
#   - schema 후보: e1%ct% (대소문자 무관)
#   - table 후보: fct_whole_op_ct 만 (정확히)
# -------------------------
finder_sql = """
SELECT table_schema, table_name
FROM information_schema.tables
WHERE table_type='BASE TABLE'
  AND table_schema ILIKE :schema_pat
  AND table_name = :table_name_exact
ORDER BY table_schema, table_name
"""

with engine.connect() as conn:
    candidates = pd.read_sql_query(
        text(finder_sql),
        conn,
        params={"schema_pat": "e1%ct%", "table_name_exact": "fct_whole_op_ct"},
    )

if candidates.empty:
    with engine.connect() as conn:
        schemas = pd.read_sql_query(
            text("""
                SELECT schema_name
                FROM information_schema.schemata
                WHERE schema_name ILIKE 'e1%'
                ORDER BY schema_name
            """),
            conn,
        )
    display(schemas)
    raise RuntimeError(
        "정확히 'fct_whole_op_ct' 테이블을 찾지 못했습니다.\n"
        "위 e1* 스키마 목록을 보고 실제 스키마명을 확인해주세요."
    )

display(candidates)

OP_SCHEMA = candidates.iloc[0]["table_schema"]
OP_TABLE  = candidates.iloc[0]["table_name"]

# 대문자 스키마 대응: 항상 쌍따옴표
op_fqn = f'"{OP_SCHEMA}"."{OP_TABLE}"'

# -------------------------
# (C) final_ct 로드 & 매핑
# 조건:
# - station = 'whole' 만
# - month = df.month
# - remark = 'PD' 또는 'Non-PD'
# -------------------------
valid_remarks = ("PD", "Non-PD")
months_needed = sorted(df["month"].dropna().unique().tolist())

sql_final_ct = f"""
SELECT
    month,
    remark,
    final_ct
FROM {op_fqn}
WHERE station = 'whole'
  AND month = ANY(:months)
  AND remark = ANY(:remarks)
"""

with engine.connect() as conn:
    df_ct = pd.read_sql_query(
        text(sql_final_ct),
        conn,
        params={"months": months_needed, "remarks": list(valid_remarks)},
    )

# (month, remark) 중복 방어: final_ct max로 1개로 정리
df_ct["month"] = df_ct["month"].astype(str)
df_ct["remark"] = df_ct["remark"].astype(str)
df_ct = df_ct.groupby(["month", "remark"], as_index=False)["final_ct"].max()

# df에 조인: remark_norm 기준으로 조인 (Non-PD 누락 방지)
df = df.merge(
    df_ct,
    how="left",
    left_on=["month", "remark_norm"],
    right_on=["month", "remark"],
)

# 불필요한 right-side remark 컬럼 정리
if "remark_y" in df.columns:
    # 원본 remark는 remark_x에 남아있으므로 제거/정리
    df.drop(columns=["remark_y"], inplace=True, errors="ignore")
if "remark_x" in df.columns:
    df.rename(columns={"remark_x": "remark"}, inplace=True)

# -------------------------
# (D) 이 DataFrame이 생성(업데이트)된 시각을 updated_at 컬럼에 추가 (KST)
#  - 모든 행 동일 timestamp
#  - file_path 바로 뒤에 삽입
# -------------------------
df_time = pd.Timestamp.now(tz="Asia/Seoul")
if "updated_at" in df.columns:
    df.drop(columns=["updated_at"], inplace=True)

# 최종 출력 컬럼 구성
out_cols = [
    "barcode_information",
    "station",
    "end_day",
    "end_time",
    "remark",
    "result",
    "goodorbad",
    "file_path",
    "final_ct",
]

df_out = df[out_cols].copy()

insert_at = df_out.columns.get_loc("file_path") + 1
df_out.insert(insert_at, "updated_at", df_time)

df_out
```

```python
# =========================
# [추가 셀] d1_machine_log.mes_fail2_wasted_time
# 1) 테이블 없으면 생성
# 2) PK를 (barcode_information, end_day, end_time) 로 설정
# 3) df_out 신규 데이터만 INSERT (중복은 SKIP: ON CONFLICT DO NOTHING)
#
# 전제: df_out 컬럼에 아래가 존재
#  - barcode_information, station, end_day, end_time, remark, result,
#    goodorbad, file_path, final_ct, updated_at(=데이터프레임 생성 시각)
# =========================

import math
import pandas as pd
from sqlalchemy import text
from psycopg2.extras import execute_values

SCHEMA_NAME = "d1_machine_log"
TABLE_NAME  = "mes_fail2_wasted_time"
FQN = f"{SCHEMA_NAME}.{TABLE_NAME}"

# -------------------------
# 0) 테이블 생성(없으면) + 기본 컬럼 보장
# -------------------------
ddl_create = f"""
CREATE SCHEMA IF NOT EXISTS {SCHEMA_NAME};

CREATE TABLE IF NOT EXISTS {FQN} (
    barcode_information  TEXT,
    end_day              TEXT,   -- YYYYMMDD
    end_time             TEXT,   -- HH:MM:SS (text)

    station              TEXT,
    remark               TEXT,
    result               TEXT,
    goodorbad            TEXT,
    file_path            TEXT,
    final_ct             NUMERIC,

    df_updated_at        TIMESTAMPTZ,  -- dataframe 생성 시각

    created_at           TIMESTAMPTZ NOT NULL DEFAULT now(),
    updated_at           TIMESTAMPTZ NOT NULL DEFAULT now()
);
"""
with engine.begin() as conn:
    conn.execute(text(ddl_create))

# -------------------------
# 1) PK를 (barcode_information, end_day, end_time) 로 맞추기
#   - 기존 PK가 있으면 제거 후 재설정
#   - PK 컬럼에 NULL 있으면 PK 설정이 실패하므로 미리 검사
# -------------------------
with engine.connect() as conn:
    null_chk = conn.execute(text(f"""
        SELECT
          SUM(CASE WHEN barcode_information IS NULL OR barcode_information='' THEN 1 ELSE 0 END) AS n_barcode_null,
          SUM(CASE WHEN end_day IS NULL OR end_day='' THEN 1 ELSE 0 END) AS n_end_day_null,
          SUM(CASE WHEN end_time IS NULL OR end_time='' THEN 1 ELSE 0 END) AS n_end_time_null
        FROM {FQN};
    """)).mappings().first()

if (null_chk["n_barcode_null"] or null_chk["n_end_day_null"] or null_chk["n_end_time_null"]):
    raise RuntimeError(
        f"PK 설정 불가: PK 컬럼에 NULL/빈값 존재\n"
        f"- barcode_information NULL/빈값: {null_chk['n_barcode_null']}\n"
        f"- end_day NULL/빈값: {null_chk['n_end_day_null']}\n"
        f"- end_time NULL/빈값: {null_chk['n_end_time_null']}\n"
        f"먼저 해당 행을 정리(삭제/보정) 후 다시 실행하세요."
    )

with engine.begin() as conn:
    # 컬럼 NOT NULL (PK 전제)
    conn.execute(text(f'ALTER TABLE {FQN} ALTER COLUMN barcode_information SET NOT NULL;'))
    conn.execute(text(f'ALTER TABLE {FQN} ALTER COLUMN end_day SET NOT NULL;'))
    conn.execute(text(f'ALTER TABLE {FQN} ALTER COLUMN end_time SET NOT NULL;'))

    # 기존 PK 있으면 드롭
    pk_name = conn.execute(text("""
        SELECT c.conname
        FROM pg_constraint c
        JOIN pg_class t ON t.oid = c.conrelid
        JOIN pg_namespace n ON n.oid = t.relnamespace
        WHERE c.contype='p'
          AND n.nspname = :schema
          AND t.relname = :table
    """), {"schema": SCHEMA_NAME, "table": TABLE_NAME}).scalar()

    if pk_name:
        conn.execute(text(f'ALTER TABLE {FQN} DROP CONSTRAINT "{pk_name}";'))

    # 새 PK 설정
    conn.execute(text(f"""
        ALTER TABLE {FQN}
        ADD CONSTRAINT pk_mes_fail2_wasted_time
        PRIMARY KEY (barcode_information, end_day, end_time);
    """))

print(f"[OK] PK set: ({'barcode_information, end_day, end_time'}) on {FQN}")

# -------------------------
# 2) df_out -> INSERT (중복 SKIP)
# -------------------------
def _py(v):
    if v is None:
        return None
    try:
        if isinstance(v, float) and math.isnan(v):
            return None
    except Exception:
        pass
    if str(v) == "NaT":
        return None
    return v

df_save = df_out.copy()

# df_out.updated_at -> df_updated_at 로 저장
if "updated_at" in df_save.columns:
    df_save["df_updated_at"] = df_save["updated_at"]
else:
    df_save["df_updated_at"] = pd.Timestamp.now(tz="Asia/Seoul")

INSERT_COLS = [
    "barcode_information",
    "end_day",
    "end_time",
    "station",
    "remark",
    "result",
    "goodorbad",
    "file_path",
    "final_ct",
    "df_updated_at",
]

for c in INSERT_COLS:
    if c not in df_save.columns:
        df_save[c] = None

df_save = df_save[INSERT_COLS]

# PK 컬럼이 비어있는 행은 저장 불가 → 제외
df_save = df_save[
    df_save["barcode_information"].notna() & (df_save["barcode_information"].astype(str).str.len() > 0) &
    df_save["end_day"].notna() & (df_save["end_day"].astype(str).str.len() > 0) &
    df_save["end_time"].notna() & (df_save["end_time"].astype(str).str.len() > 0)
].copy()

records = [tuple(_py(v) for v in row) for row in df_save.itertuples(index=False, name=None)]
if not records:
    print("[INFO] insert 대상 레코드가 없습니다(df_out 비었거나 PK 컬럼이 비어있음).")
else:
    cols_sql = ", ".join(INSERT_COLS)

    sql_ins = f"""
    INSERT INTO {FQN} ({cols_sql})
    VALUES %s
    ON CONFLICT (barcode_information, end_day, end_time) DO NOTHING;
    """

    with engine.begin() as conn:
        raw = conn.connection  # psycopg2 connection
        with raw.cursor() as cur:
            execute_values(
                cur,
                sql_ins,
                records,
                template=f"({', '.join(['%s'] * len(INSERT_COLS))})",
                page_size=2000,
            )

    print(f"[OK] attempted insert rows: {len(records)} (duplicates skipped) into {FQN}")
```
